```python
import requests                 
import nltk                      
from bs4 import BeautifulSoup   
from collections import Counter
import matplotlib.pyplot as plt
```

```python
r = requests.get("https://www.gutenberg.org/files/16/16-h/16-h.htm")

r.encoding = 'utf-8'

html = r.text
```

```python
soup = BeautifulSoup(html)
text = soup.text
```

```python
tokenizer = nltk.tokenize.RegexpTokenizer("\w+")
tokens = tokenizer.tokenize(text)
words = [token.lower() for token in tokens]
```

```python
nltk.download('stopwords')
stop_words = nltk.corpus.stopwords.words("english")
```

```python
words_clean = [word for word in words if word not in stop_words]
count = Counter(words_clean)
top_ten = count.most_common(10)
```

```python
print(top_ten)
```

```python
words = [word for word, count in top_ten]
counts = [count for word, count in top_ten]

#Plot a histogram
plt.figure(figsize=(10, 6))
plt.bar(words, counts, color='skyblue')

plt.title('Top 10 Most Common Meaningful Words', fontsize=16)
plt.xlabel('Words', fontsize=12)
plt.ylabel('Frequency', fontsize=12)

plt.xticks(rotation=45)
plt.tight_layout()
plt.show() 
```
